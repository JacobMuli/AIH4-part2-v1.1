```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

%matplotlib inline
```

```python
# UPDATE THESE IF YOU UPLOAD TO /content INSTEAD
ADM_PATH = "/content/adm_meru_potato.csv"
HV_PATH = "/content/hvstat_meru_potato.csv"

def load_csv(path):
    if os.path.exists(path):
        print("Loading:", path)
        return pd.read_csv(path)
    else:
        raise FileNotFoundError("File not found — upload or correct path: " + path)

adm = load_csv(ADM_PATH)
hv = load_csv(HV_PATH)

adm.columns = adm.columns.str.lower()
hv.columns = hv.columns.str.lower()

print("ADM shape:", adm.shape)
print("HVSTAT shape:", hv.shape)
```

```python
print("ADM sample rows:")
display(adm.head())

print("\nHVSTAT sample rows:")
display(hv.head())
```

```python
def extract_year(df):
    if "harvest_year" in df.columns:
        return df["harvest_year"]
    if "period_date" in df.columns:
        return pd.to_datetime(df["period_date"], errors="ignore").dt.year
    if "start_date" in df.columns:
        return pd.to_datetime(df["start_date"], errors="ignore").dt.year
    return None

adm["year"] = extract_year(adm)
hv["year"] = extract_year(hv)

print(adm[["year"]].head())
print(hv[["year"]].head())
```

```python
# Yield over years (HVSTAT — main dataset with yield available)
if "yield" in hv.columns:
    hv_year = hv.groupby("year")["yield"].mean()
    plt.figure(figsize=(10,4))
    plt.plot(hv_year.index, hv_year.values, marker='o')
    plt.title("HVSTAT: Meru Potato Yield over Time")
    plt.xlabel("Year")
    plt.ylabel("Yield (t/ha)")
    plt.grid(True)
    plt.show()

# Area vs Production
if "area" in hv.columns and "production" in hv.columns:
    plt.figure(figsize=(6,4))
    plt.scatter(hv["area"], hv["production"])
    plt.title("HVSTAT: Area vs Production")
    plt.xlabel("Area (ha)")
    plt.ylabel("Production (tonnes)")
    plt.grid(True)
    plt.show()

# Yield by Sub-County
if "admin_2" in hv.columns:
    sub = hv.groupby("admin_2")["yield"].mean().sort_values()
    plt.figure(figsize=(12,4))
    sub.plot(kind="bar")
    plt.title("HVSTAT: Yield by Sub-county (Meru)")
    plt.ylabel("Yield (t/ha)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
print("Summary statistics (HVSTAT only):")
display(hv[["year","area","production","yield"]].describe())
```

```python
# Group Meru by year and sum area + production
grouped = hv.groupby("year").agg({
    "area": "sum",
    "production": "sum"
}).reset_index()

grouped["yield"] = grouped["production"] / grouped["area"]
grouped = grouped.sort_values("year").reset_index(drop=True)

print("Modeling dataset:")
display(grouped)
```

```python
df = grouped.copy()

# Lag features
for lag in [1,2,3]:
    df[f"yield_lag_{lag}"] = df["yield"].shift(lag)

# Rolling mean
df["yield_roll3"] = df["yield"].rolling(3).mean().shift(1)

# Drop rows missing lags
df = df.dropna().reset_index(drop=True)

print("Features added:")
display(df)
```

```python
# ---- Train/Test Split ----

# df contains: year, area, production, yield, yield_lag_1...lag_3, yield_roll3
# We created df in feature engineering step

FEATURES = ["yield_lag_1", "yield_lag_2", "yield_lag_3", "yield_roll3"]

# Hold out last year for testing
train = df.iloc[:-1]
test = df.iloc[-1:]

X_train = train[FEATURES]
y_train = train["yield"]

X_test = test[FEATURES]
y_test = test["yield"]

print("Training years:", train["year"].tolist())
print("Testing year:", test["year"].tolist())
```

```python
# Train LR
lr = LinearRegression()
lr.fit(X_train, y_train)

# Train Random Forest
rf = RandomForestRegressor(n_estimators=300, random_state=42)
rf.fit(X_train, y_train)

# Predictions
lr_pred = lr.predict(X_test)[0]
rf_pred = rf.predict(X_test)[0]

# Metrics function
def metrics(true, pred):
    mae = mean_absolute_error([true], [pred])
    mse = mean_squared_error([true], [pred])
    rmse = np.sqrt(mse)  # Calculate RMSE manually
    mape = float(np.abs((true - pred) / true) * 100)

    return {
        "MAE": mae,
        "RMSE": rmse,
        "MAPE_%": mape
    }

print("---- Model Performance ----")
print("Linear Regression:", metrics(y_test, lr_pred))
print("Random Forest:", metrics(y_test, rf_pred))

print("\nActual yield:", float(y_test))
print("LR predicted yield:", float(lr_pred))
print("RF predicted yield:", float(rf_pred))
```

```python
# ---- Recursive multi-year forecasting function ----

import datetime

def recursive_forecast(grouped, rf_model, final_year=None):
    """
    grouped: dataframe with historical 'year', 'yield', 'area'
    rf_model: trained RandomForest model
    final_year: target year to roll predictions up to
    """

    # Last historical year in the dataset
    last_year = int(grouped['year'].iloc[-1])

    # If no final year specified → predict only next year
    if final_year is None:
        final_year = last_year + 1

    df_hist = grouped.copy().reset_index(drop=True)
    forecast_results = []

    for year in range(last_year + 1, final_year + 1):

        # Build lag features from latest values
        y_lag_1 = df_hist['yield'].iloc[-1]
        y_lag_2 = df_hist['yield'].iloc[-2] if len(df_hist) >= 2 else y_lag_1
        y_lag_3 = df_hist['yield'].iloc[-3] if len(df_hist) >= 3 else y_lag_1
        y_roll3 = df_hist['yield'].iloc[-3:].mean()

        X_fore = np.array([[y_lag_1, y_lag_2, y_lag_3, y_roll3]])

        pred_yield = rf_model.predict(X_fore)[0]
        area = df_hist['area'].iloc[-1]   # last known area
        pred_tonnes = pred_yield * area

        forecast_results.append({
            "year": year,
            "predicted_yield_t_ha": pred_yield,
            "predicted_tonnage": pred_tonnes
        })

        # Append prediction for next iteration
        df_hist = pd.concat([
            df_hist,
            pd.DataFrame({"year":[year], "yield":[pred_yield], "area":[area], "production":[pred_tonnes]})
        ], ignore_index=True)

    return pd.DataFrame(forecast_results)
```

```python
# ---- Multi-year recursive forecasting ----

last_year = int(grouped['year'].max())
current_year = datetime.datetime.now().year

print("Dataset last year:", last_year)
print("Forecasting up to:", current_year)

forecast_df = recursive_forecast(grouped, rf, final_year=current_year)

print("\nRecursive multi-year forecast:")
display(forecast_df)

# Final year prediction (current year)
final_pred_yield = forecast_df.iloc[-1]['predicted_yield_t_ha']
final_pred_tonnes = forecast_df.iloc[-1]['predicted_tonnage']

print("\nFinal predicted yield (t/ha):", final_pred_yield)
print("Final predicted production (tonnes):", final_pred_tonnes)
```

```python
# ---- Cold Storage Packing Engine ----

def pack_storage(total_tonnes, chamber_sizes=[1000, 500, 250], max_fill_rate=0.90):
    required_capacity = int(np.ceil(total_tonnes / max_fill_rate))

    allocation = []
    remaining = required_capacity

    for size in chamber_sizes:
        count = remaining // size
        if count > 0:
            allocation.append({"size": size, "count": int(count)})
            remaining -= size * count

    if remaining > 0:
        # allocate one more smallest chamber
        allocation.append({"size": chamber_sizes[-1], "count": 1})

    total_capacity = sum(a["size"] * a["count"] for a in allocation)
    utilization = total_tonnes / total_capacity

    return {
        "predicted_tonnes": total_tonnes,
        "required_capacity": required_capacity,
        "allocated_capacity": total_capacity,
        "allocation": allocation,
        "utilization_rate": utilization
    }

storage_plan = pack_storage(final_pred_tonnes)

print("---- Storage Recommendation ----")
storage_plan
```

```python
import joblib
import json

# Save models
joblib.dump(rf, "/content/rf_meru_yield_model.joblib")
joblib.dump(lr, "/content/lr_meru_yield_model.joblib")

# Save forecast + storage plan
results = {
    "rf_forecast_yield_t_ha": float(final_pred_yield),
    "lr_forecast_yield_t_ha": float(lr_pred),
    "predicted_production_tonnes": float(final_pred_tonnes),
    "storage_recommendation": storage_plan
}

with open("/content/meru_yield_results.json", "w") as f:
    json.dump(results, f, indent=4)

print("Saved files:")
print(" - /content/rf_meru_yield_model.joblib")
print(" - /content/lr_meru_yield_model.joblib")
print(" - /content/meru_yield_results.json")
```
